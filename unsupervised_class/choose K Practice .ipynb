```python
import numpy as np
from kmeans import plot_k_means, get_simple_data, cost
import matplotlib.pyplot as plt

def main():
    X=get_simple_data()
    plt.scatter(X[:,0],X[:,1])
    plt.show()
    costs=np.empty(10)
    costs[0]=None
    for k in range(1,10):
        M,R=plot_k_means(X,k,show_plots=False)
        costs[k]=cost(X,R,M)
    plt.plot(costs)
    plt.title("Costs Vs K ")
    plt.show()
main()
```
